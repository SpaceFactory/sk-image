```python
# Import statements
from skimage import data, io, filters

# save a copy of the coin image as "image"
image = data.coins()
# add "image" to plotting queue
io.imshow(image)
io.show()
```

```python
# ... or any other NumPy array!
# create a copy of "image after applying the sorbel edge finding filter
edges = filters.sobel(image)
io.imshow(edges)
io.show()
```
